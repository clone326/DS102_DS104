# Filtering Companies with 100 or more entries

```python
# Filtering by company with more than or equal to 100 entries
df_company_100 = df_data_science_salaries_cleaned.groupby('company').filter(lambda x : len(x) >= 100)

#Double checking
df_company_100.groupby(['company'])['timestamp'].count()
```
